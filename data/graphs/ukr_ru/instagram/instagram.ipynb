```python
"""
Directed mention graph: account → mentioned account.
Vectorized with pandas explode — handles 1.3M rows fast.
"""

import ast, json, numpy as np, pandas as pd, torch
from torch_geometric.data import Data
from collections import defaultdict


def _parse_list(val):
    if isinstance(val, list): return val
    if not isinstance(val, str): return []
    try:
        r = ast.literal_eval(val)
        return r if isinstance(r, list) else []
    except: return []

def _parse_dict(val):
    if isinstance(val, dict): return val
    if not isinstance(val, str): return {}
    try: return ast.literal_eval(val)
    except: return {}


def build_mention_graph(df, text_embeddings=None, embed_dim=768):
    """
    Args:
        df: posts dataframe with handle, mentions, statistics, account, 
            subscriberCount, score, type, languageCode, likeAndViewCountsDisabled
        text_embeddings: optional dict {handle: np.array(768,)}
    Returns:
        PyG Data, handle_to_idx dict, handle list
    """
    df = df.copy()
    df["handle"] = df["handle"].str.lower().str.strip()
    df["_mentions"] = df["mentions"].apply(_parse_list).apply(
        lambda x: [m.lower().strip() for m in x if m.strip()]
    )

    # ── Edges (vectorized) ──
    edges = (
        df[["handle", "_mentions", "score"]]
        .explode("_mentions")
        .dropna(subset=["_mentions"])
        .rename(columns={"handle": "src", "_mentions": "dst"})
    )
    print(f"Raw edges: {len(edges):,}")

    # aggregate parallel edges
    edge_agg = edges.groupby(["src", "dst"]).agg(
        count=("score", "size"),
        avg_score=("score", "mean"),
    ).reset_index()

    # ── Node index ──
    all_handles = sorted(set(df["handle"]) | set(edges["dst"]))
    h2i = {h: i for i, h in enumerate(all_handles)}
    N = len(all_handles)

    edge_index = torch.from_numpy(np.stack([
        edge_agg["src"].map(h2i).values,
        edge_agg["dst"].map(h2i).values,
    ]).astype(np.int64))

    edge_attr = torch.tensor(
        edge_agg[["count", "avg_score"]].values, dtype=torch.float
    )

    # ── Node features (16d structural) ──
    df["_stats"] = df["statistics"].apply(_parse_dict)
    df["_acc"] = df["account"].apply(_parse_dict)
    df["_al"] = df["_stats"].apply(lambda s: (s.get("actual") or {}).get("favoriteCount", 0) or 0)
    df["_ac"] = df["_stats"].apply(lambda s: (s.get("actual") or {}).get("commentCount", 0) or 0)
    df["_el"] = df["_stats"].apply(lambda s: (s.get("expected") or {}).get("favoriteCount", 0) or 0)
    df["_ec"] = df["_stats"].apply(lambda s: (s.get("expected") or {}).get("commentCount", 0) or 0)
    df["_verified"] = df["_acc"].apply(lambda a: int(a.get("verified", False)))

    agg = df.groupby("handle").agg(
        subs=("subscriberCount", "max"),
        verified=("_verified", "max"),
        n=("handle", "size"),
        al=("_al", "mean"), ac=("_ac", "mean"),
        el=("_el", "mean"), ec=("_ec", "mean"),
        score=("score", "mean"),
        al_sum=("_al", "sum"), el_sum=("_el", "sum"),
        fv=("type", lambda x: (x == "video").mean()),
        fp=("type", lambda x: (x == "photo").mean()),
        fa=("type", lambda x: (x == "album").mean()),
        nlang=("languageCode", "nunique"),
        fdis=("likeAndViewCountsDisabled", "mean"),
    )
    agg["overperf"] = np.where(agg["el_sum"] > 0, agg["al_sum"] / agg["el_sum"], 1.0)

    # degree
    in_deg = edge_agg.groupby("dst").size().to_dict()
    out_deg = edge_agg.groupby("src").size().to_dict()

    x = torch.zeros((N, 16), dtype=torch.float)
    for h, i in h2i.items():
        if h in agg.index:
            r = agg.loc[h]
            x[i] = torch.tensor([
                np.log1p(r.subs), float(r.verified), np.log1p(r.n),
                np.log1p(r.al), np.log1p(r.ac),
                np.log1p(r.el), np.log1p(r.ec),
                r.score, r.overperf,
                r.fv, r.fp, r.fa,
                float(r.nlang), r.fdis,
                np.log1p(in_deg.get(h, 0)), np.log1p(out_deg.get(h, 0)),
            ])
        else:
            x[i, 14] = np.log1p(in_deg.get(h, 0))
            x[i, 15] = np.log1p(out_deg.get(h, 0))

    # ── Text embeddings ──
    if text_embeddings:
        x_text = torch.zeros((N, embed_dim), dtype=torch.float)
        for h, i in h2i.items():
            if h in text_embeddings:
                x_text[i] = torch.as_tensor(text_embeddings[h], dtype=torch.float)
        x = torch.cat([x, x_text], dim=1)
        print(f"Text embeddings matched: {sum(1 for h in h2i if h in text_embeddings):,}/{N:,}")

    data = Data(x=x, edge_index=edge_index, edge_attr=edge_attr, num_nodes=N)
    data.node_handles = all_handles
    data.num_structural_features = 16

    print(f"Graph: {N:,} nodes, {edge_index.shape[1]:,} edges, {x.shape[1]}d features")
    return data, h2i, all_handles
```

```python
data, h2i, all_handles = build_mention_graph(df)

torch.save({
    "data": data,
    "h2i": h2i,
    "handles": all_handles,
}, "mention_graph.pt")

# later:
# ckpt = torch.load("mention_graph.pt")
# data, h2i, handles = ckpt["data"], ckpt["h2i"], ckpt["handles"]
```

```python
def graph_mem(data):
    total = 0
    for key, val in data:
        if isinstance(val, torch.Tensor):
            total += val.element_size() * val.nelement()
    mb = total / 1024**2
    gb = total / 1024**3
    print(f"x:          {data.x.shape} → {data.x.element_size() * data.x.nelement() / 1024**2:.1f} MB")
    print(f"edge_index: {data.edge_index.shape} → {data.edge_index.element_size() * data.edge_index.nelement() / 1024**2:.1f} MB")
    print(f"edge_attr:  {data.edge_attr.shape} → {data.edge_attr.element_size() * data.edge_attr.nelement() / 1024**2:.1f} MB")
    print(f"Total:      {mb:.1f} MB ({gb:.2f} GB)")

graph_mem(data)
```
